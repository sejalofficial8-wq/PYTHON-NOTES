# NUMERIC DATATYPE

```python
# INTERGER
a=10
print(a)
print(type(a))
```

```python
a=10.0
print(a)
print(type(a))
```

```python
# STRING- ORDERED SEQUENCE OF CHARACTERS
# SINGLE LINE DENOTED BY ' ' OR " " AND
# MULTI-LINE STRING DENOTED BY ''' '''

# INDEXED & IMMUTUABLE
s=""    #empty string 
print(s)
print(type(s))
```

```python
s="decode"
print(s)
print(type(s))
print(s[0])
s[0]="z"
print(s)
```

```python
s='''hello
   welcome
    to bhopal.'''
print(s)
```

```python
# LIST- Collection of element
# it is denoted by [] with csv.
# indexed,ordered,mutuable and heterogenous in nature
l=[]    # empty list
print(l)
print(type(l))
```

```python
l=[10,20,30,"decode",10.5]
print(l)
print(l[3])    #indexed
l[0]="bhopal"
print(l)
```

```python
# TUPLE
#COLLECTION OF ELEMENT
# DENOTED BY () WITH CSV
#ORDERED,INDEXED,IMMUTUABLE AND HETEROGENEOUS
t=()
print(t)
print(type(t))
```

```python
# single value tuple
t=('10.5',)
print(t)
print(type(t))
```

```python
t=(10,20,300,"decode",10.5)
print(t)
print(t[2])
t[2]=400
print(t)
```

```python
#DICTIONARY
# COLLECTION OF DATAVALUE
# IT is denoted by {k:v} with key:value pair
# ordered,unindexed and mutuable.
#keys...unique
#values--- duplicate
d={}      #empty dictionary  dict
print(d)
print(type(d))
```

```python
d={"name":"abc","subject":"DA","Fees":"25000"}
print(d)
print(d["name"])
d["name"]="def"
print(d)
```

```python
d={}
d["name"]="def"
print(d)
```

```python
d={"name":["def","abc","pqr"],"subject":["DA","DA","DS"]}
print(d)
print(d["name"])
```

```python
# SET- COLLECTION OF ELEMENT
#it is denoted by {} with csv
# unordered,unindexed,mutuable,heterogeneousin nature
# stored unique value.
s=set()
print(s)
print(type(s))
```

```python
s={10,10,10,20,10,100,200,"decode","h"}
print(s)
print(type(s))
s.add(1000)
print(s)
```
