```python
#Q] WAPP THAT DETERMINES THE TYPE OF TRIANGLE BASED ON THE THE LENGTH
#OF ITS 3 SIDES. TYPES OF TRIANGLE: EQUILATERAL,ISOSCELES,SCALEN
s1=int(input("enter side 1: "))
s2=int(input("enter side 2: "))
s3=int(input("enter side 3: "))
sum_of_3sides=s1+s2+s3
if s1+s2>s3 and s2+s3>s1 and s3+s1>s2:
    if s1==s2 and s2==s3 and s3==s1:
      print("it is an eqilateral triangle")
    elif s1==s2 and s2==s3 and s3==s1:
      print ("it is an isoceles triangle")
    else:
       print("it is a scalen triangle")
else:
     print("the given sides does not form a triangle")
```
