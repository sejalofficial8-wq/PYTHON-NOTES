```python
# Q] WAPP TO PRINT REVERSE OF ANY GIVEN STRING?
s=input("enter any string: ")
rev=""
for i in s:
    rev=i+rev
print(rev) 
print(s[ : :-1])
```

```python
#Q] WAPP TO CHECK WHETHER A GIVEN STRING IS PALINDROME OR NOT?
s=input("enter any string: ")
rev=""
for i in s:
    rev=i+rev
if s==rev:
    print(s,"palindrome")
else:
    print(s,"not a palindrome")
```

```python
#SAME QUESTION WITH OTHER METHOD
s=input("enter any string: ")
if s==s[ : :-1]:
    print(s,"is a palindrome")
else:
    print(s,"not a palindrome")
  
```

## METHODS



```python
#upper()- strobj.upper()  [upper case]
#lower()- strobj.lower()  [lower case]
#title()- strobj.title()  [converts first letter capital of each word]
#capitalize()- strobj.capitalize()  [converts only first letter capital]
s="welcome to bhopal"
print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
```

```python
#isalpha()    # isalphabet
#strobj.isalpha()
s="decode"
s1="decode1"
print(s.isalpha())
print(s1.isalpha())
```

```python
#isdigit()
#strobj.isdigit()
s="1234"
s1="decode1234"
print(s.isdigit())
print(s1.isdigit())
```

```python
#isalnum()  # isalphabet & numeric
#strobj.isalnum()
s="decode123"
print(s.isalnum())
```

```python
#count()- return occurence of particular character/substring in a given string.
#strobj.count(string/character,[start],[stop])
s="welcome to bhopal"
print(s.count("o"))
print(s.count("o",5))
print(s.count("o",5,10))
print(s.count("o",5,9))
print(s.count("decode"))
print(s.count("bhopal"))
```

```python
#strip()
#strobj.strip(string)
s="...rrdecodedtt,&,,/"
print(s.strip(".,r/&te"))
```

```python
#index()- return position
# strobj.index(char/substring,[start],[stop])
s="welcome to bhopal"
print(s.index("o"))
print(s.index("o",5))
print(s.index("bhopal"))
print(s.index("o",5,9))
```

```python
#find()- return position
#strobj.find(char/substring,[start],[stop])
s="welcome to bhopal"
print(s.find("o"))
print(s.find("o",5))
print(s.find("bhopal"))
print(s.find("o",5,9))
```

```python
#replace()
#strobj.replace(oldstring,newstring,[count])
s="welcome to bhopal"
print(s.replace("bhopal","decode"))
print(s.replace("o","a"))
```

```python
#split()
#strobj.split([separator],[maxsplit])
#by default- whitespace,  maxsplit- max no of separator
# split will always return list of strings.
s="welcome to city of lakes"
print(s.split())
print(s.split(" ",4))
print(s.split(" "))
```

```python
s="welcome,to,city of lakes"
print(s.split(",",1))
```
