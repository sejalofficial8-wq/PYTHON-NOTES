# OPERATORS


## ARITHMETIC OPERATOR

```python
# Arithmetic operator
a=10
b=2
print ("addition : ", a+b)
print ("subtraction : ", a-b)
print ("multiplication: ",a*b)
print ("simple division : ",a/b)
print ("floor division : ",a//b)
print ("modulus : ", a%b)
print ("exponential power : ",a**b)
```

# Assignment operator 

```python
# Assignment operator =,+=,-=,*=,/*,//=,**==
a=10
b= 8
c=a+b
a+=b  # a=a+b
a-=10 # a=a-10
a*=5  # a=a*5
a//=5  #a=a//5
b*=10  #b=b*10
print(a)
print(b)
print(c)
```

## COMPARISON OPERATOR

```python
# COMPARISON OPERATOR ==, !=,>=,<=,>,<
a=10
b=8
print(a==b)
print(a!=b)
print(a>=b)
print(a<=b)
print(a>b)
print(a<b)
```

## LOGICAL OPERATOR

```python
#LOGICAL OPERATOR
a=10
b=8
c=12
print(a>b and b>c)
print(a>b or b>c)
print(a>b and not b>c)
```

# IDENTITY OPERATOR

```python
## IDENTITY OPERATOR is,is not
# id() - returns the memory address of an object
a=10
b=10
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```

# MEMBERSHIP OPERATOR

```python
# MEMBERSHIP OPERATOR in, not in
a="hello"
print("he" in a)
print("h" not in a)
l=[10,20,"hello"]
print("h" in l)
```

# BITWISE OPERATOR

```python
# BITWISE
# & - bitwise and (&-ampercent)
# | - bitwise or (|-pipesign)
# ^ - bitwise xor (^- cap)
# bin()
a=10
b=8
print(a,bin(a))
print(b,"",bin(b))
print(bin(a&b))
print(a&b)
print()
print(bin(a|b))
print(a|b)
print()
print(bin(a^b))
print(a^b)
```
