```python
t=(10,20,30,10.5,100)
l=list(t)
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]>l[j]:
            l[i],l[j]=l[j],l[i]
t1=tuple(l)
print(t1)
```
